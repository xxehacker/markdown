```python
import matplotlib.pyplot as plt
import pandas as pd

# Actual Tesla Quarterly Revenue Data
revenue_data = {
    'Date': ['2023-03-31', '2023-06-30', '2023-09-30', '2023-12-31'],
    'Revenue': [23.329, 24.927, 23.350, 25.167]  # in billions
}

# Actual Tesla Stock Price Data
stock_data = {
    'Date': ['2024-03-15', '2024-03-14', '2024-03-13', '2024-03-12'],  # Sample recent dates
    'Close': [163.57, 162.50, 160.76, 165.18]  # Sample closing prices
}

# Convert data into pandas DataFrames
df_revenue = pd.DataFrame(revenue_data)
df_stock = pd.DataFrame(stock_data)

# Convert 'Date' to datetime and set as index
df_revenue['Date'] = pd.to_datetime(df_revenue['Date'])
df_revenue.set_index('Date', inplace=True)

df_stock['Date'] = pd.to_datetime(df_stock['Date'])
df_stock.set_index('Date', inplace=True)

# Create subplots
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot stock data
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price (USD)', color='tab:red')
ax1.plot(df_stock.index, df_stock['Close'], color='tab:red', marker='o')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for the revenue data
ax2 = ax1.twinx()
ax2.set_ylabel('Revenue (Billion USD)', color='tab:blue')
ax2.plot(df_revenue.index, df_revenue['Revenue'], color='tab:blue', marker='o')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title and layout
plt.title('Tesla Stock Price and Revenue Dashboard')
fig.tight_layout()

# Show plot
plt.show()
```
